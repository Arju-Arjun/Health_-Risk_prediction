# **Feature Engineering**


```python
df["BMI"] = df["Weight"] / (df["Height"] ** 2)
df = df.drop(columns=["Weight", "Height"])
df
```

```python
for i in df.columns:
  if df[i].dtype=='object':
    print(i,':',df[i].unique())
```

```python
# df = pd.get_dummies(df, columns=["Gender"], drop_first=True)
df['Gender']=df['Gender'].map({'Female':0,'Male':1})
```

```python
map_activity={'Sedentary': 0, 'Lightly Active': 1, 'Moderately Active': 2, 'Active': 3}
df["Activity Level"] = df["Activity Level"].map(map_activity)
risk_mapping = {'Low': 0, 'Medium': 1, 'High': 2, 'Very High': 3, 'Severe': 4}
df['Health Risk Level'] = df["Health Risk Level"].map(risk_mapping)
```

# **Outlier Detection**

```python
sns.boxplot(df['Body Fat Percentage'])
```

```python
Q1 = df['Body Fat Percentage'].quantile(0.25)
Q3 = df['Body Fat Percentage'].quantile(0.75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df = df[(df['Body Fat Percentage'] >= lower_bound) & (df['Body Fat Percentage'] <= upper_bound)]
```

# **Initial Modeling**

Logistic Regression, Decision Tree, Random Forest, XGBoost, SVM.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,confusion_matrix
```

```python
x=df.drop(columns='Health Risk Level')
y=df['Health Risk Level']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
print(f'{x_train.shape}\n{x_test.shape}\n{y_train.shape}\n{y_test.shape}')
```

```python
models = {
    "Linear Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    'XGBClassifier':XGBClassifier(),
    "SVC":SVC()
}

```

```python
def evaluate_model(y_true, y_pred):

    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="weighted")  # Fix for multiclass
    recall = recall_score(y_true, y_pred, average="weighted")  # Fix for multiclass
    f1 = f1_score(y_true, y_pred, average="weighted")

    return accuracy,precision,recall,f1
```

```python
results = {}
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python

for name, model in models.items():

    if name == "Logistic Regression":  # Apply scaled data for Logistic Regression
        model.fit(x_train_scaled, y_train)
        y_pred = model.predict(x_test_scaled)
    else:
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)

    accuracy, precision, recall, f1 = evaluate_model(y_test, y_pred)
    results[name] = {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}
```

```python
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
for i in range(15,25):
      rf=RandomForestClassifier( n_estimators=250,        # Number of trees
                                criterion="gini", # Loss function
                                max_depth=i,            # Maximum depth of each tree
                                min_samples_split=4,     # Minimum samples required to split a node
                                min_samples_leaf=2,      # Minimum samples required in a leaf node
                                max_features="sqrt",     # Number of features to consider for best split
                                bootstrap=True,          # Use bootstrap sampling
                                random_state=42,         # Ensure reproducibility

                                )
      rf.fit(x_train, y_train)
      y_pred = rf.predict(x_test)
      print(f'accuracy {i}: {accuracy_score(y_test, y_pred)}')

```
